```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
df=pd.read_csv("/content/StudentsPerformance.csv")
```

```python
import matplotlib.pyplot as plt

# Calculate average scores by gender
avg_scores_gender = df.groupby('gender').mean()

# Create bar plot
avg_scores_gender.plot(kind='bar', figsize=(8, 6))
plt.title('Average Scores by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Score')
plt.xticks(rotation=0)
plt.legend(loc='upper center')
plt.show()
```

```python
# Create count plot
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='test preparation course', hue='gender')
plt.title('Test Preparation Course Participation by Gender')
plt.xlabel('Test Preparation Course')
plt.ylabel('Count')
plt.legend(title='Gender')
plt.show()
```

```python
# Create scatter plot
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='math score', y='reading score', hue='gender')
plt.title('Math vs. Reading Scores')
plt.xlabel('Math Score')
plt.ylabel('Reading Score')
plt.legend(title='Gender')
plt.grid(True)
plt.show()
```

```python
import seaborn as sns

# Create box plot
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='race/ethnicity', y='math score', hue='gender')
plt.title('Math Score Distribution by Race/Ethnicity')
plt.xlabel('Race/Ethnicity')
plt.ylabel('Math Score')
plt.legend(title='Gender')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='test preparation course', hue='gender')
plt.title('Test Preparation Course Completion by Gender')
plt.xlabel('Test Preparation Course')
plt.ylabel('Count')
plt.legend(title='Gender')
plt.show()
```
